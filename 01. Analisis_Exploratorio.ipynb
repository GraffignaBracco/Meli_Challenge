## Preguntas de Negocio

1. ¿Cuál es el precio promedio por categoría?
2. ¿Cuantas son las cantidades vendidas promedio por categoría?

```python
df.groupby('category_id').aggregate({'price': 'mean', 'sold_quantity': 'mean' })
```
